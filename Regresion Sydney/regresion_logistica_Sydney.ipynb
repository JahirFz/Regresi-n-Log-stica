```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as special
from scipy.optimize import curve_fit
import seaborn as sns
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#carga de archivo csv
df = pd.read_csv('Datos_Limpios_Sydney.csv')
```

```python
#Rellenamos nulos
df=df.fillna(method='bfill')
df=df.fillna(method='ffill')
```

```python
#Convirtiendo una variable a dicotómica
df['room_type'] = df['room_type'].replace(
    [ 'Private room', 'Shared room'],
    'Hotel room'
)
df['room_type'] = np.where(df['room_type'] == 'Entire home/apt', 1, 0)

#Convirtiendo una variable a dicotómica
df['host_response_time'] = df['host_response_time'].replace([ 'No definido', 'a few days or more', 'within an hour'],'within a few hours')
df['host_response_time'] = np.where(df['host_response_time'] == 'within a day', 1, 0)
```

HOST_IS_SUPERHOST VS NUMBER_OF_REVIEWS, BEDS, PRICE

```python
#Declaramos las varibles dependientes e independientes para la regresión Logística
Vars_Indep=df[['number_of_reviews', 'price', 'beds']]
var_Dep=df['host_is_superhost']
```

```python
#Redefinimos las varibles
X = Vars_Indep
y = var_Dep
```

```python
#Dividimos el conjunto de datos en la parte de entrenamiento80 y prueba20:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)
```

```python
#Se escalan todos los datos
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las varibles "X" tanto de entrenamiento como de prueba, utilizaremos
X_train1 = escalar.fit_transform(X_train)
X_test1 = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Relizamos una predicción
y_pred = algoritmo.predict(X_test)
y_pred
```

```python
#Verifico la matriz de Confunsión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión')
print(matriz)
```

```python
#Calculo la precisión del modelo
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred, average='binary', pos_label='t')
print('Precisión del modelo cierto en HOST_IS_SUPERHOST VS NUMBER_OF_REVIEWS, BEDS, PRICE')
print(precision)

precision = precision_score(y_test, y_pred, average='binary', pos_label='f')
print('Precisión del modelo falso en HOST_IS_SUPERHOST VS NUMBER_OF_REVIEWS, BEDS, PRICE')
print(precision)
```

```python
#Se escalan todos los datos
escalar = StandardScaler()
```

```python
#Para realizar el escalamiento de las varibles "X" tanto de entrenamiento como de prueba, utilizaremos
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Relizamos una predicción
y_pred = algoritmo.predict(X_test)
y_pred
```

```python
#Verifico la matriz de Confunsión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión')
print(matriz)
```

```python
#Calculo la precisión del modelo
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred, average='binary', pos_label='t')
print('Precisión del modelo verdadero en host_is_superhost vs room_type, bedrooms')
print(precision)

precision = precision_score(y_test, y_pred, average='binary', pos_label='f')
print('Precisión del modelo falso en host_is_superhost vs room_type, bedrooms')
print(precision)
```

```python
#Calculado la exactitud del modelo
from sklearn.metrics import accuracy_score

exactitud = accuracy_score(y_test, y_pred)
print('Exactitud del modelo en host_is_superhost vs room_type, bedrooms')
print(exactitud)
```

```python
#Calculo la sensibilidad del modelo
from sklearn.metrics import recall_score

sensibilidad = recall_score(y_test, y_pred, average='binary', pos_label='t')
print('Sensibilidad del modelo verdadero en host_is_superhost vs room_type, bedrooms')
print(sensibilidad)

sensibilidad = recall_score(y_test, y_pred, average='binary', pos_label='f')
print('Sensibilidad del modelo falso en host_is_superhost vs room_type, bedrooms')
print(sensibilidad)
```
